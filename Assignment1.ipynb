##############################################################################

The hypothesis function for linear regression is given by:

$h_{\theta}(x) = \theta^T x$

The Mean Squared Error (MSE) cost function is:

$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2$

### Derivation of the Gradient of J($\theta$) with respect to $\theta$

Let's expand the cost function in terms of individual components. For a single training example $(x^{(i)}, y^{(i)})$,

we know that,

Error is the diffrence of prediction and the actual

The error is:

$e^{(i)}$ = prediction - actual

$e^{(i)} = h_{\theta}(x^{(i)}) - y^{(i)} = \theta^T x^{(i)} - y^{(i)}$

We want to find the partial derivative of $J(\theta)$ with respect to each parameter $\theta_j$:

$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{\partial}{\partial \theta_j} \left[ \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2 \right]$

Using the chain rule, for a single term in the sum:

$\frac{\partial}{\partial \theta_j} (h_{\theta}(x^{(i)}) - y^{(i)})^2 = 2 (h_{\theta}(x^{(i)}) - y^{(i)}) \frac{\partial}{\partial \theta_j} (h_{\theta}(x^{(i)}) - y^{(i)})$

Now, let's find the derivative of $(h_{\theta}(x^{(i)}) - y^{(i)})$ with respect to $\theta_j$:

$\frac{\partial}{\partial \theta_j} (\theta^T x^{(i)} - y^{(i)}) = \frac{\partial}{\partial \theta_j} \left( \sum_{k=0}^{d} \theta_k x_k^{(i)} - y^{(i)} \right)$

Since $x_j^{(i)}$ is the only term that depends on $\theta_j$ in the sum:

$= x_j^{(i)}$

Substituting this back:

$\frac{\partial}{\partial \theta_j} (h_{\theta}(x^{(i)}) - y^{(i)})^2 = 2 (h_{\theta}(x^{(i)}) - y^{(i)}) x_j^{(i)}$

Now, we can write the full partial derivative of the cost function:

$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{2m} \sum_{i=1}^{m} 2 (h_{\theta}(x^{(i)}) - y^{(i)}) x_j^{(i)}$

$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)}) x_j^{(i)}$

In vector form, the gradient of $J(\theta)$ with respect to $\theta$ is:

$\nabla_{\theta} J(\theta) = \frac{1}{m} X^T (X\theta - \vec{y})$

Where:
* $X$ is the design matrix (m x (d+1))
* $\theta$ is the parameter vector ((d+1) x 1)
* $\vec{y}$ is the target vector (m x 1)

### Gradient Descent Update Rule for $\theta$

The gradient descent update rule for each parameter $\theta_j$ is given by:

$\theta_j := \theta_j - \alpha \frac{\partial J(\theta)}{\partial \theta_j}$

Substituting the derived gradient:

$\theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)}) x_j^{(i)}$

In vector form, the update rule for the entire parameter vector $\theta$ is:

$\theta (t+1) :=  (t) \theta - \alpha \frac{1}{m} X^T (X\theta - \vec{y})$

Where $\alpha$ is the learning rate, which controls the size of the steps taken during gradient descent.

### Derivation of the Gradient of $J(\theta)$ with respect to $\theta$

Let's expand the cost function in terms of individual components. For a single training example $(x^{(i)}, y^{(i)})$, the error is:

$e^{(i)} = h_{\theta}(x^{(i)}) - y^{(i)} = \theta^T x^{(i)} - y^{(i)}$

We want to find the partial derivative of $J(\theta)$ with respect to each parameter $\theta_j$:

$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{\partial}{\partial \theta_j} \left[ \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)})^2 \right]$

Using the chain rule:

$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{2m} \sum_{i=1}^{m} 2 \cdot (h_{\theta}(x^{(i)}) - y^{(i)}) \cdot \frac{\partial}{\partial \theta_j} (h_{\theta}(x^{(i)}) - y^{(i)})$

Since $h_{\theta}(x^{(i)}) = \theta_0 x_0^{(i)} + \theta_1 x_1^{(i)} + \dots + \theta_j x_j^{(i)} + \dots$, the derivative with respect to $\theta_j$ is:

$\frac{\partial}{\partial \theta_j} (\theta^T x^{(i)} - y^{(i)}) = x_j^{(i)}$

Substituting this back into the summation:

$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)}) x_j^{(i)}$

In vector notation, the gradient $\nabla_{\theta} J(\theta)$ is:

$\nabla_{\theta} J(\theta) = \frac{1}{m} X^T (X\theta - y)$

### Gradient Descent Update Rule

The update rule for gradient descent is to subtract the gradient scaled by the learning rate $\alpha$:

$\theta := \theta - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)}) - y^{(i)}) x^{(i)}$

### (c) Normal Equation for Linear Regression

Using vector notation, express the normal equation: $\theta = (X^T X)^{-1} X^T y$. Show the derivation by minimizing $J(\theta)$.

##############################################################################

### Derivation of the Normal Equation

We represent the cost function in matrix form:

$J(\theta) = \frac{1}{2m} (X\theta - y)^T (X\theta - y)$

Expanding the quadratic form:

$J(\theta) = \frac{1}{2m} ((X\theta)^T - y^T)(X\theta - y)$
$J(\theta) = \frac{1}{2m} (\theta^T X^T X \theta - \theta^T X^T y - y^T X \theta + y^T y)$

Since $\theta^T X^T y$ is a scalar ($1 \times 1$ matrix), it is equal to its own transpose $(y^T X \theta)$. We can combine them:

$J(\theta) = \frac{1}{2m} (\theta^T X^T X \theta - 2(X^T y)^T \theta + y^T y)$

To find the minimum, we take the derivative with respect to the vector $\theta$ and set it to zero:

$\nabla_{\theta} J(\theta) = \frac{1}{2m} (2 X^T X \theta - 2 X^T y) = 0$

Solving for $\theta$:

$X^T X \theta = X^T y$
$\theta = (X^T X)^{-1} X^T y$

```python
# 1. Setup Data
# Ensure X and Y are numpy arrays for vectorization
X_arr = np.array(X)  # Shape: (10000, 3)
Y_arr = np.array(Y)  # Shape: (10000,)
m, n = X_arr.shape

# Initialize parameters
theta_sgd = np.zeros(n)
alpha = 0.01  # From part (b)
epochs = 100
sgd_costs = []

# 2. Training Loop
for epoch in range(epochs):
    # Step A: Shuffle indices at the start of each epoch
    indices = np.random.permutation(m)

    for i in indices:
        xi = X_arr[i]  # Select one row (feature vector)
        yi = Y_arr[i]  # Select corresponding target

        # Step B: Vectorized Prediction (Dot Product)
        # hi = theta0*x0 + theta1*x1 + theta2*x2
        hi = np.dot(xi, theta_sgd)

        # Step C: Immediate Update
        # Gradient for one example is (h(x) - y) * x_vector
        error = hi - yi
        gradient = error * xi
        theta_sgd -= alpha * gradient

    # Step D: Record Cost after each epoch to monitor convergence
    # We use the vectorized Mean Squared Error formula
    predictions = np.dot(X_arr, theta_sgd)
    cost = np.sum((predictions - Y_arr)**2) / (2 * m)
    sgd_costs.append(cost)

# 3. Visualization
plt.figure(figsize=(10, 6))
plt.plot(sgd_costs)
plt.title("SGD Convergence (Vectorized)")
plt.xlabel("Epoch")
plt.ylabel("Cost J(theta)")
plt.grid(True)
plt.show()

print(f"Final Theta: {theta_sgd}")
```

```python
#mini-batch GD
X_arr = np.array(X)  # (10000, 3)
Y_arr = np.array(Y)  # (10000,)
m, n = X_arr.shape

theta_mb = np.zeros(n)
alpha = 0.01
batch_size = 64
epochs = 100
mb_costs = []

# 2. Training Loop
for epoch in range(epochs):
    # Step A: Shuffle data at the start of each epoch
    indices = np.random.permutation(m)
    X_shuffled = X_arr[indices]
    Y_shuffled = Y_arr[indices]

    # Step B: Iterate through data in steps of 64
    for i in range(0, m, batch_size):
        # Extract the mini-batch
        X_batch = X_shuffled[i : i + batch_size]
        Y_batch = Y_shuffled[i : i + batch_size]

        # Step C: Vectorized Prediction for the whole batch
        # h_batch shape: (64,)
        h_batch = np.dot(X_batch, theta_mb)

        # Step D: Calculate Gradient for the batch
        # (X_batch.T @ error) gives the sum of gradients for all 64 points
        error = h_batch - Y_batch
        gradient = np.dot(X_batch.T, error) / X_batch.shape[0]

        # Step E: Update parameters
        theta_mb -= alpha * gradient

    # Track cost per epoch
    current_preds = np.dot(X_arr, theta_mb)
    cost = np.mean((current_preds - Y_arr)**2) / 2
    mb_costs.append(cost)

# 3. Visualization
plt.plot(mb_costs)
plt.title("Mini-Batch GD (Size 64) Convergence")
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.show()

print(f"Final Theta: {theta_mb}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

# --- 1. Data Preparation ---
# Using the already loaded X and Y from previous cells (Housing_dataset.csv)

# Convert global X and Y (lists) to numpy arrays for split and calculations
X_arr = np.array(X)
Y_arr = np.array(Y)

# Split into training and validation sets (80/20)
X_train, X_val, y_train, y_val = train_test_split(X_arr, Y_arr, test_size=0.2, random_state=42)

# Reshape y_train and y_val to be column vectors for matrix operations
y_train = y_train.reshape(-1, 1)
y_val = y_val.reshape(-1, 1)

train_size = len(X_train) # Get the actual training size after split

# --- 2. Normal Equation (Exact Solution) ---
theta_normal = np.linalg.inv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)

# --- 3. Optimization Parameters ---
alpha = 0.01
epochs = 100 # Increased epochs for better convergence visibility across methods
batch_size = 64

def compute_mse(X_data, y_data, theta_param):
    # Ensure y_data is a column vector for this calculation if theta_param is also a column vector
    return np.mean((X_data.dot(theta_param) - y_data) ** 2) / 2 # MSE definition was (1/2m)sum(error^2)

# --- 4. Batch Gradient Descent ---
theta_bgd = np.zeros((X_train.shape[1], 1)) # Initialize theta as column vector
bgd_costs = []
for i in range(epochs):
    gradients = (1/train_size) * X_train.T.dot(X_train.dot(theta_bgd) - y_train)
    theta_bgd -= alpha * gradients
    bgd_costs.append(compute_mse(X_train, y_train, theta_bgd))

# --- 5. Stochastic Gradient Descent (SGD) ---
theta_sgd = np.zeros((X_train.shape[1], 1)) # Initialize theta as column vector
sgd_costs = []
for epoch in range(epochs):
    indices = np.random.permutation(train_size)
    for i in indices:
        xi = X_train[i:i+1] # Keep as 2D array for dot product
        yi = y_train[i:i+1] # Keep as 2D array for dot product
        gradient = xi.T.dot(xi.dot(theta_sgd) - yi)
        theta_sgd -= alpha * gradient
    # Track cost after each epoch
    sgd_costs.append(compute_mse(X_train, y_train, theta_sgd))


# --- 6. Mini-batch Gradient Descent ---
theta_mini = np.zeros((X_train.shape[1], 1)) # Initialize theta as column vector
mini_costs = []
for epoch in range(epochs):
    indices = np.random.permutation(train_size)
    X_s, y_s = X_train[indices], y_train[indices] # Shuffled data
    for i in range(0, train_size, batch_size):
        xi = X_s[i:i+batch_size]
        yi = y_s[i:i+batch_size]
        gradient = (1/len(yi)) * xi.T.dot(xi.dot(theta_mini) - yi)
        theta_mini -= alpha * gradient
    # Track cost after each epoch
    mini_costs.append(compute_mse(X_train, y_train, theta_mini))

# --- 7. Validation & Comparison ---
results = {
    "Normal Equation": compute_mse(X_val, y_val, theta_normal),
    "Batch GD": compute_mse(X_val, y_val, theta_bgd),
    "SGD": compute_mse(X_val, y_val, theta_sgd),
    "Mini-batch GD": compute_mse(X_val, y_val, theta_mini)
}

print("Validation MSE Results:")
for method, mse in results.items():
    print(f"{method}: {mse:.6f}")

# --- 8. Plotting ---
plt.figure(figsize=(10, 5))
plt.plot(bgd_costs, label='Batch GD', linewidth=2)
plt.plot(sgd_costs, label='SGD', alpha=0.6) # x-axis should naturally align with epochs
plt.plot(mini_costs, label='Mini-batch', linestyle='--')
plt.yscale('log')
plt.xlabel('Epochs') # Changed x-axis label for clarity
plt.ylabel('Cost (MSE)')
plt.title('Convergence: Batch vs SGD vs Mini-batch')
plt.legend()
plt.grid(True)
plt.show()
```

## 1. Convexity of the Logistic Regression Loss

The empirical risk for logistic regression is given by:

$$J(\theta) = -\frac{1}{n} \sum_{i=1}^{n} \left[ y^{(i)} \log(\sigma(\theta^T x^{(i)})) + (1 - y^{(i)}) \log(1 - \sigma(\theta^T x^{(i)})) \right]$$

where the sigmoid function is: $\sigma(z) = \frac{1}{1+e^{-z}}$.

### (a) Derivation of the Hessian Matrix $H = \nabla^2 J(\theta)$

**Step 1: First Derivative (Gradient)** Let $h^{(i)} = \sigma(\theta^T x^{(i)})$. Using the property $\frac{\partial \sigma(z)}{\partial z} = \sigma(z)(1 - \sigma(z))$, we find the partial derivative with respect to $\theta_j$:

$$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{n} \sum_{i=1}^{n} (h^{(i)} - y^{(i)}) x_j^{(i)}$$

In vector form, the gradient is:
$$\nabla J(\theta) = \frac{1}{n} \sum_{i=1}^{n} (\sigma(\theta^T x^{(i)}) - y^{(i)}) x^{(i)}$$

**Step 2: Second Derivative (Hessian)** The Hessian $H$ is the Jacobian of the gradient. We differentiate the gradient with respect to $\theta^T$:

$$H_{jk} = \frac{\partial}{\partial \theta_k} \left[ \frac{1}{n} \sum_{i=1}^{n} (h^{(i)} - y^{(i)}) x_j^{(i)} \right]$$
$$H_{jk} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{\partial h^{(i)}}{\partial \theta_k} \right) x_j^{(i)}$$

Since $\frac{\partial h^{(i)}}{\partial \theta_k} = h^{(i)}(1 - h^{(i)}) x_k^{(i)}$, we get:
$$H_{jk} = \frac{1}{n} \sum_{i=1}^{n} h^{(i)}(1 - h^{(i)}) x_j^{(i)} x_k^{(i)}$$

In matrix form:
$$H = \frac{1}{n} \sum_{i=1}^{n} \sigma(\theta^T x^{(i)})(1 - \sigma(\theta^T x^{(i)})) x^{(i)} (x^{(i)})^T$$

### (b) Proof that $H$ is Positive Semi-Definite (PSD)

To prove $H$ is PSD, we must show $v^T H v \geq 0$ for any vector $v$:

$$v^T H v = v^T \left[ \frac{1}{n} \sum_{i=1}^{n} h^{(i)}(1-h^{(i)}) x^{(i)} (x^{(i)})^T \right] v$$
$$v^T H v = \frac{1}{n} \sum_{i=1}^{n} h^{(i)}(1-h^{(i)}) (v^T x^{(i)}) (x^{(i)T} v)$$
$$v^T H v = \frac{1}{n} \sum_{i=1}^{n} \ h^{(i)}(1-h^{(i)})(v^T x^{(i)})^2_{\geq 0}$$

Since $0 < \sigma(z) < 1$, the term $h^{(i)}(1-h^{(i)})$ is always positive. The term $(v^T x^{(i)})^2$ is a square, which is always $\geq 0$. Thus, $v^T H v \geq 0$, confirming $H$ is PSD and $J(\theta)$ is convex.

```python
import numpy as np
import matplotlib.pyplot as plt


def add_intercept(x):
    """Add intercept to matrix x.

    Args:
        x: 2D NumPy array.

    Returns:
        New matrix same as x with 1's in the 0th column.
    """
    new_x = np.zeros((x.shape[0], x.shape[1] + 1), dtype=x.dtype)
    new_x[:, 0] = 1
    new_x[:, 1:] = x

    return new_x


def load_dataset(csv_path, label_col='y', add_intercept=False):
    """Load dataset from a CSV file.

    Args:
         csv_path: Path to CSV file containing dataset.
         label_col: Name of column to use as labels (should be 'y' or 'l').
         add_intercept: Add an intercept entry to x-values.

    Returns:
        xs: Numpy array of x-values (inputs).
        ys: Numpy array of y-values (labels).
    """

    def add_intercept_fn(x):
        global add_intercept
        return add_intercept(x)

    # Validate label_col argument
    allowed_label_cols = ('y', 't')
    if label_col not in allowed_label_cols:
        raise ValueError('Invalid label_col: {} (expected {})'
                         .format(label_col, allowed_label_cols))

    # Load headers
    with open(csv_path, 'r') as csv_fh:
        headers = csv_fh.readline().strip().split(',')

    # Load features and labels
    x_cols = [i for i in range(len(headers)) if headers[i].startswith('x')]
    l_cols = [i for i in range(len(headers)) if headers[i] == label_col]
    inputs = np.loadtxt(csv_path, delimiter=',', skiprows=1, usecols=x_cols)
    labels = np.loadtxt(csv_path, delimiter=',', skiprows=1, usecols=l_cols)

    if inputs.ndim == 1:
        inputs = np.expand_dims(inputs, -1)

    if add_intercept:
        inputs = add_intercept_fn(inputs)

    return inputs, labels


def plot(x, y, theta, save_path, correction=1.0):
    """Plot dataset and fitted logistic regression parameters.

    Args:
        x: Matrix of training examples, one per row.
        y: Vector of labels in {0, 1}.
        theta: Vector of parameters for logistic regression model.
        save_path: Path to save the plot.
        correction: Correction factor to apply (Problem 2(e) only).
    """
    # Plot dataset
    plt.figure()
    plt.plot(x[y == 1, -2], x[y == 1, -1], 'bx', linewidth=2)
    plt.plot(x[y == 0, -2], x[y == 0, -1], 'go', linewidth=2)

    # Plot decision boundary (found by solving for theta^T x = 0)
    x1 = np.arange(min(x[:, -2]), max(x[:, -2]), 0.01)
    x2 = -(theta[0] / theta[2] + theta[1] / theta[2] * x1
           + np.log((2 - correction) / correction) / theta[2])
    plt.plot(x1, x2, c='red', linewidth=2)
    plt.xlim(x[:, -2].min()-.1, x[:, -2].max()+.1)
    plt.ylim(x[:, -1].min()-.1, x[:, -1].max()+.1)

    # Add labels and save to disk
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.savefig(save_path)


class LogisticRegression:
    """Logistic regression with Newton's Method as the solver.

    Example usage:
        > clf = LogisticRegression()
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """
    def __init__(self, step_size=0.01, max_iter=10000, eps=1e-5,
                 theta_0=None, verbose=True):
        """
        Args:
            step_size: Step size for iterative solvers only.
            max_iter: Maximum number of iterations for the solver.
            eps: Threshold for determining convergence.
            theta_0: Initial guess for theta. If None, use the zero vector.
            verbose: Print loss values during training.
        """
        self.theta = theta_0
        self.step_size = step_size
        self.max_iter = max_iter
        self.eps = eps
        self.verbose = verbose

    def fit(self, x, y):
        n, d = x.shape
        if self.theta is None:
            self.theta = np.zeros(d)

        for i in range(self.max_iter):
            theta_old = np.copy(self.theta)

            # Calculate probabilities
            z = x.dot(self.theta)
            h = 1 / (1 + np.exp(-z))

            # Gradient
            grad = (1/n) * x.T.dot(h - y)

            # Hessian
            diag_weight = (1/n) * h * (1 - h)
            hess = (x.T * diag_weight).dot(x)

            # Newton update
            self.theta -= np.linalg.inv(hess).dot(grad)

            # Check convergence
            if np.linalg.norm(self.theta - theta_old, 1) < self.eps:
                print(f'Converged at iteration {i}')
                break


    def predict(self, x):
        """Make a prediction given new inputs x.

        Args:
            x: Inputs of shape (n_examples, dim).

        Returns:
            Outputs of shape (n_examples,).
        """
        # *** START CODE HERE ***
        # For logistic regression, predictions are probabilities sigma(theta^T * x)
        return 1 / (1 + np.exp(-np.dot(x, self.theta)))
        # *** END CODE HERE ***

def main(train_path, valid_path, save_path):
    """Problem: Logistic regression with Newton's Method.

    Args:
        train_path: Path to CSV file containing dataset for training.
        valid_path: Path to CSV file containing dataset for validation.
        save_path: Path to save predicted probabilities using np.savetxt().
    """
    # Note: util.load_dataset adds the intercept (column of 1s) for us
    x_train, y_train = load_dataset(train_path, add_intercept=True)

    # *** START CODE HERE ***
    # Train a logistic regression classifier
    clf = LogisticRegression()
    clf.fit(x_train, y_train)

    # Plot decision boundary on top of validation set
    x_valid, y_valid = load_dataset(valid_path, add_intercept=True)
    plot(x_valid, y_valid, clf.theta, save_path.replace('.txt', '.png'))

    # Use np.savetxt to save predictions on eval set to save_path
    p_valid = clf.predict(x_valid)
    np.savetxt(save_path, p_valid)
    # *** END CODE HERE ***


if __name__ == '__main__':
    main(train_path='ds1_train.csv',
         valid_path='ds1_valid.csv',
         save_path='logreg_pred_1.txt')

    main(train_path='ds2_train.csv',
         valid_path='ds2_valid.csv',
         save_path='logreg_pred_2.txt')
```

### 3. Linear Decision Boundary of Gaussian Classifiers

#### Generative Model Assumptions
Assume the following generative model:
* $p(y = 1) = \pi$
* $p(y = 0) = 1 - \pi$
* $p(x | y = k) = \mathcal{N}(x; \mu_k, \Sigma) = \frac{1}{(2\pi)^{d/2}|\Sigma|^{1/2}} \exp\left(-\frac{1}{2}(x-\mu_k)^T\Sigma^{-1}(x-\mu_k)\right)$ for $k \in \{0, 1\}$

---

#### 1. Bayes' Rule Expansion
Using Bayes' Rule, the posterior probability of $y=1$ given $x$ is:
$$p(y=1|x) = \frac{p(x|y=1)p(y=1)}{p(x|y=1)p(y=1) + p(x|y=0)p(y=0)}$$

To simplify this into the logistic form $\frac{1}{1 + e^{-z}}$, we divide both the numerator and denominator by $p(x|y=1)p(y=1)$:
$$p(y=1|x) = \frac{1}{1 + \frac{p(x|y=0)p(y=0)}{p(x|y=1)p(y=1)}} = \frac{1}{1 + \exp\left(-\log\left(\frac{p(x|y=1)p(y=1)}{p(x|y=0)p(y=0)}\right)\right)}$$

---

#### 2. Evaluating the Log-Ratio
Let $z = \log\left(\frac{p(x|y=1)p(y=1)}{p(x|y=0)p(y=0)}\right)$. Expanding the log of the product:
$$z = \log\left(\frac{p(x|y=1)}{p(x|y=0)}\right) + \log\left(\frac{\pi}{1-\pi}\right)$$

Substitute the multivariate Gaussian densities. Note that the constants $\frac{1}{(2\pi)^{d/2}|\Sigma|^{1/2}}$ cancel out because the covariance matrix $\Sigma$ is shared between classes:
$$z = \left[ -\frac{1}{2}(x-\mu_1)^T\Sigma^{-1}(x-\mu_1) + \frac{1}{2}(x-\mu_0)^T\Sigma^{-1}(x-\mu_0) \right] + \log\left(\frac{\pi}{1-\pi}\right)$$

---

#### 3. Algebraic Simplification
Expand the quadratic terms using $(x-\mu_k)^T\Sigma^{-1}(x-\mu_k) = x^T\Sigma^{-1}x - x^T\Sigma^{-1}\mu_k - \mu_k^T\Sigma^{-1}x + \mu_k^T\Sigma^{-1}\mu_k$:
$$z = -\frac{1}{2}\left(x^T\Sigma^{-1}x - 2\mu_1^T\Sigma^{-1}x + \mu_1^T\Sigma^{-1}\mu_1\right) + \frac{1}{2}\left(x^T\Sigma^{-1}x - 2\mu_0^T\Sigma^{-1}x + \mu_0^T\Sigma^{-1}\mu_0\right) + \log\left(\frac{\pi}{1-\pi}\right)$$

Notice that the quadratic term $x^T\Sigma^{-1}x$ cancels out, which results in a linear decision boundary:
$$z = \mu_1^T\Sigma^{-1}x - \mu_0^T\Sigma^{-1}x - \frac{1}{2}\mu_1^T\Sigma^{-1}\mu_1 + \frac{1}{2}\mu_0^T\Sigma^{-1}\mu_0 + \log\left(\frac{\pi}{1-\pi}\right)$$

Factoring out $x$:
$$z = \underbrace{(\mu_1 - \mu_0)^T\Sigma^{-1}}_{\theta^T} x + \underbrace{\frac{1}{2}\mu_0^T \Sigma^{-1} \mu_0 - \frac{1}{2}\mu_1^T \Sigma^{-1} \mu_1 + \log \frac{\pi}{1-\pi}}_{\theta_0}$$

---

#### 4. Final Form
The expression simplifies to the logistic sigmoid function:
$$p(y=1|x) = \frac{1}{1 + \exp(-(\theta^T x + \theta_0))}$$

Where the parameters are defined in terms of $\pi, \mu_0, \mu_1,$ and $\Sigma$ as:
* $\theta = \Sigma^{-1}(\mu_1 - \mu_0)$ (noting that $\Sigma^{-1}$ is symmetric)
* $\theta_0 = \frac{1}{2}\mu_0^T \Sigma^{-1} \mu_0 - \frac{1}{2}\mu_1^T \Sigma^{-1} \mu_1 + \log \frac{\pi}{1-\pi}$


### 4. Maximum Likelihood Estimation Derivation

To derive the Maximum Likelihood Estimates (MLE), we maximize the log-likelihood of the joint distribution of the observed data $\{(x^{(i)}, y^{(i)})\}_{i=1}^n$.

#### 1. The Log-Likelihood Function
The joint likelihood is the product of the probabilities of each independent example:
$$\mathcal{L}(\pi, \mu_0, \mu_1, \Sigma) = \prod_{i=1}^{n} p(x^{(i)}, y^{(i)}) = \prod_{i=1}^{n} p(x^{(i)} | y^{(i)}) p(y^{(i)})$$

Taking the natural logarithm gives the log-likelihood $\ell$:
$$\ell = \sum_{i=1}^{n} \log p(x^{(i)} | y^{(i)}) + \sum_{i=1}^{n} \log p(y^{(i)})$$

Substituting the Gaussian density for $p(x|y)$ and Bernoulli for $p(y)$:
$$\ell = \sum_{i=1}^{n} \left[ -\frac{d}{2}\log(2\pi) - \frac{1}{2} \log |\Sigma| - \frac{1}{2} (x^{(i)} - \mu_{y^{(i)}})^T \Sigma^{-1} (x^{(i)} - \mu_{y^{(i)}}) \right] + \sum_{i=1}^{n} [y^{(i)} \log \pi + (1 - y^{(i)}) \log (1 - \pi)]$$

---

#### 2. Deriving $\pi$
To find the estimate for the class prior $\pi$, we take the partial derivative of $\ell$ with respect to $\pi$ and set it to zero:
$$\frac{\partial \ell}{\partial \pi} = \sum_{i=1}^{n} \left[ \frac{y^{(i)}}{\pi} - \frac{1 - y^{(i)}}{1 - \pi} \right] = 0$$

Multiplying by $\pi(1 - \pi)$:
$$\sum_{i=1}^{n} [y^{(i)}(1 - \pi) - (1 - y^{(i)})\pi] = 0$$
$$\sum_{i=1}^{n} [y^{(i)} - y^{(i)}\pi - \pi + y^{(i)}\pi] = 0 \implies \sum_{i=1}^n y^{(i)} - n\pi = 0$$
$$\pi = \frac{1}{n} \sum_{i=1}^{n} \mathbb{I}\{y^{(i)} = 1\}$$

---

#### 3. Deriving $\mu_k$
To find $\mu_k$ (for $k \in \{0, 1\}$), we differentiate $\ell$ with respect to $\mu_k$. Only the terms where $y^{(i)} = k$ remain:
$$\nabla_{\mu_k} \ell = \sum_{i: y^{(i)}=k} \nabla_{\mu_k} \left[ -\frac{1}{2} (x^{(i)} - \mu_k)^T \Sigma^{-1} (x^{(i)} - \mu_k) \right]$$

Using the matrix gradient identity $\nabla_z (z - a)^T W (z - a) = 2W(z - a)$, we set the gradient to zero:
$$\sum_{i: y^{(i)}=k} \Sigma^{-1} (x^{(i)} - \mu_k) = 0$$

Since $\Sigma$ is positive definite, we can multiply by $\Sigma$ and rearrange:
$$\sum_{i=1}^n \mathbb{I}\{y^{(i)} = k\} (x^{(i)} - \mu_k) = 0$$
$$\sum_{i=1}^n \mathbb{I}\{y^{(i)} = k\} x^{(i)} = \sum_{i=1}^n \mathbb{I}\{y^{(i)} = k\} \mu_k$$
$$\mu_k = \frac{\sum_{i=1}^n \mathbb{I}\{y^{(i)} = k\} x^{(i)}}{\sum_{i=1}^n \mathbb{I}\{y^{(i)} = k\}}$$

---

#### 4. Deriving $\Sigma$
The covariance $\Sigma$ is shared across both classes. Ignoring constants with respect to $\Sigma$, we maximize:
$$\ell(\Sigma) = -\frac{n}{2} \log |\Sigma| - \frac{1}{2} \sum_{i=1}^{n} (x^{(i)} - \mu_{y^{(i)}})^T \Sigma^{-1} (x^{(i)} - \mu_{y^{(i)}})$$



Using the property $(x-a)^T \Sigma^{-1} (x-a) = \text{tr}((x-a)(x-a)^T \Sigma^{-1})$, let $S_i = (x^{(i)} - \mu_{y^{(i)}})(x^{(i)} - \mu_{y^{(i)}})^T$:
$$\ell(\Sigma) = -\frac{n}{2} \log |\Sigma| - \frac{1}{2} \text{tr}\left( \left( \sum_{i=1}^n S_i \right) \Sigma^{-1} \right)$$

Taking the derivative with respect to $\Sigma^{-1}$ (where $\frac{\partial \log |A|}{\partial A^{-1}} = -A$) and setting to zero:
$$\frac{n}{2} \Sigma - \frac{1}{2} \sum_{i=1}^n S_i = 0$$
$$\Sigma = \frac{1}{n} \sum_{i=1}^{n} (x^{(i)} - \mu_{y^{(i)}}) (x^{(i)} - \mu_{y^{(i)}})^T$$

```python
# These are the helper functions in the file
import matplotlib.pyplot as plt
import numpy as np


def add_intercept(x):
    """Add intercept to matrix x.

    Args:
        x: 2D NumPy array.

    Returns:
        New matrix same as x with 1's in the 0th column.
    """
    new_x = np.zeros((x.shape[0], x.shape[1] + 1), dtype=x.dtype)
    new_x[:, 0] = 1
    new_x[:, 1:] = x

    return new_x


def load_dataset(csv_path, label_col='y', add_intercept=False):
    """Load dataset from a CSV file.

    Args:
         csv_path: Path to CSV file containing dataset.
         label_col: Name of column to use as labels (should be 'y' or 'l').
         add_intercept: Add an intercept entry to x-values.

    Returns:
        xs: Numpy array of x-values (inputs).
        ys: Numpy array of y-values (labels).
    """

    def add_intercept_fn(x):
        global add_intercept
        return add_intercept(x)

    # Validate label_col argument
    allowed_label_cols = ('y', 't')
    if label_col not in allowed_label_cols:
        raise ValueError('Invalid label_col: {} (expected {})'
                         .format(label_col, allowed_label_cols))

    # Load headers
    with open(csv_path, 'r') as csv_fh:
        headers = csv_fh.readline().strip().split(',')

    # Load features and labels
    x_cols = [i for i in range(len(headers)) if headers[i].startswith('x')]
    l_cols = [i for i in range(len(headers)) if headers[i] == label_col]
    inputs = np.loadtxt(csv_path, delimiter=',', skiprows=1, usecols=x_cols)
    labels = np.loadtxt(csv_path, delimiter=',', skiprows=1, usecols=l_cols)

    if inputs.ndim == 1:
        inputs = np.expand_dims(inputs, -1)

    if add_intercept:
        inputs = add_intercept_fn(inputs)

    return inputs, labels


def plot(x, y, theta, save_path, correction=1.0):
    """Plot dataset and fitted logistic regression parameters.

    Args:
        x: Matrix of training examples, one per row.
        y: Vector of labels in {0, 1}.
        theta: Vector of parameters for logistic regression model.
        save_path: Path to save the plot.
        correction: Correction factor to apply (Problem 2(e) only).
    """
    # Plot dataset
    plt.figure()
    plt.plot(x[y == 1, -2], x[y == 1, -1], 'bx', linewidth=2)
    plt.plot(x[y == 0, -2], x[y == 0, -1], 'go', linewidth=2)

    # Plot decision boundary (found by solving for theta^T x = 0)
    x1 = np.arange(min(x[:, -2])-4, max(x[:, -2])+4, 0.01)
    x2 = -(theta[0] / theta[2] + theta[1] / theta[2] * x1
           + np.log((2 - correction) / correction) / theta[2])
    plt.plot(x1, x2, c='red', linewidth=2)
    plt.xlim(x[:, -2].min()-4, x[:, -2].max()+4)
    plt.ylim(x[:, -1].min()-4, x[:, -1].max()+4)

    # Add labels and save to disk
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.savefig(save_path)

class LogisticRegression:
    """Logistic regression with Newton's Method as the solver."""

    def __init__(self, step_size=0.01, max_iter=100, eps=1e-5,
                 theta_0=None, verbose=True):
        self.theta = theta_0
        self.step_size = step_size
        self.max_iter = max_iter
        self.eps = eps
        self.verbose = verbose

    def fit(self, x, y):
        """Run Newton's Method to minimize the cost function."""
        m, n = x.shape
        if self.theta is None:
            self.theta = np.zeros(n)

        for i in range(self.max_iter):
            theta_old = np.copy(self.theta)

            # h(x) = sigmoid(theta^T * x)
            h = 1 / (1 + np.exp(-np.dot(x, self.theta)))

            # Gradient: grad = (1/m) * X^T * (h - y)
            gradient = (1/m) * np.dot(x.T, (h - y))

            # Hessian: H = (1/m) * X^T * diag(h(1-h)) * X
            # Optimized vectorization:
            D = h * (1 - h)
            hessian = (1/m) * np.dot(x.T, x * D[:, np.newaxis])

            # Newton's update: theta = theta - H^-1 * gradient
            self.theta -= np.linalg.solve(hessian, gradient)

            # Convergence check
            if np.linalg.norm(self.theta - theta_old, ord=1) < self.eps:
                break

    def predict(self, x):
        """Return predicted probabilities for features x."""
        return 1 / (1 + np.exp(-np.dot(x, self.theta)))
```

### 3. Bayes Rule Warm-up
Show that for any example $i$, $P(t^{(i)} = 1 | y^{(i)} = 1, x^{(i)}) = 1$.

**Proof:**
By Bayes' Rule:
$$P(t=1 | y=1, x) = \frac{P(y=1 | t=1, x) P(t=1 | x)}{P(y=1 | x)}$$

Using the Law of Total Probability, we can expand the denominator $P(y=1 | x)$ over the possible values of the true label $t \in \{0, 1\}$:
$$P(y=1 | x) = P(y=1 | t=1, x)P(t=1 | x) + P(y=1 | t=0, x)P(t=0 | x)$$

Based on the problem setup (Positive-Unlabeled learning), an example can only be labeled positive ($y=1$) if its true label is positive ($t=1$). Therefore, $P(y=1 | t=0, x) = 0$. This simplifies the denominator to:
$$P(y=1 | x) = P(y=1 | t=1, x)P(t=1 | x)$$

Substituting this back into the Bayes' Rule equation:
$$P(t=1 | y=1, x) = \frac{P(y=1 | t=1, x) P(t=1 | x)}{P(y=1 | t=1, x)P(t=1 | x)} = 1$$

Thus, if an example is observed as positive, the probability that its true label is positive is exactly 1.

---

### 4. Relating True and Observed Probabilities
Show that for any $x$, $P(t = 1 | x) = \frac{1}{\beta} P(y = 1 | x)$.

We assume the **Selected Completely At Random (SCAR)** assumption, which states that the probability of a positive example being labeled is independent of its features $x$:
$$P(y=1 | t=1, x) = P(y=1 | t=1) = \beta$$

From our derivation in Part 3, we established:
$$P(y=1 | x) = P(y=1 | t=1, x)P(t=1 | x)$$

Substituting $\beta$ into the equation:
$$P(y=1 | x) = \beta \cdot P(t=1 | x)$$

Rearranging to solve for the true probability $P(t=1 | x)$:
$$P(t = 1 | x) = \frac{1}{\beta} P(y = 1 | x)$$

**Recovery Explanation:**
A classifier trained on observed labels $y$ predicts $h(x) \approx P(y=1|x)$. Since $\beta$ represents the fraction of positive examples that were actually labeled, $P(y=1|x)$ will systematically underestimate the true probability $P(t=1|x)$. To recover the true probability, we simply rescale the classifier's output by dividing it by the constant $\beta$.

---

### 5. Estimating $\beta$
 Show that $\beta = E[h(x) | y = 1]$.

We assume that the true process is noise-free, meaning $P(t=1 | x) \in \{0, 1\}$, and that our function $h(x)$ perfectly predicts $P(y=1|x)$.

From the relationship derived in Part 4:
$$h(x) = P(y=1|x) = \beta \cdot P(t=1|x)$$

Consider the expectation of $h(x)$ conditioned on the observed label being positive ($y=1$):
$$E[h(x) | y=1] = E[\beta \cdot P(t=1|x) | y=1]$$

From Part 3, we know that if $y=1$, then the true label $t$ must be 1. Under the noise-free assumption, if $t=1$, then $P(t=1|x)$ must be 1. Thus, for all points where $y=1$, the expression simplifies:
$$E[h(x) | y=1] = E[\beta \cdot (1) | y=1]$$

Since $\beta$ is a constant:
$$E[h(x) | y=1] = \beta$$

**Sample Average Estimation:**
This result justifies estimating $\beta$ using the validation set $V$. By taking the subset of positively labeled examples $V_+ = \{x^{(i)} \in V \mid y^{(i)} = 1\}$, the empirical mean of the predicted probabilities serves as an unbiased estimator for $\beta$:
$$\beta \approx \frac{1}{|V_+|} \sum_{x^{(i)} \in V_+} h(x^{(i)})$$

## Problem 4: Linear Regression Beyond Linear Functions
Derive Objective Function and GD Rule (Problem 1)

### Subtask:
Formally show that polynomial regression (cubic) is equivalent to linear regression with a transformed feature map. Then, write down the objective function J(theta) for linear regression on the transformed dataset and its batch gradient descent update rule.

#### Instructions
1. Start by defining the cubic polynomial hypothesis $h_{\theta}(x) = \theta_0 + \theta_1x + \theta_2x^2 + \theta_3x^3$.
2. Introduce the feature map $\phi(x) = [1, x, x^2, x^3]^T$ and demonstrate how substituting this into the linear regression hypothesis $h_{\theta}(\tilde{x}) = \theta^T\tilde{x}$ (where $\tilde{x} = \phi(x)$) makes the cubic polynomial regression equivalent to linear regression.
3. Write down the objective function $J(\theta)$ for linear regression using the Mean Squared Error (MSE) on the transformed dataset ${(\tilde{x}^{(i)},y^{(i)})}_{i=1}^n$, similar to the derivation in Problem 1(a) of the notebook.
4. Write down the batch gradient descent update rule for the parameter vector $\theta$ for this objective function, again referencing the general form derived in Problem 1(a).

---

### Derivation:

**1. Cubic Polynomial Hypothesis:**
The cubic polynomial hypothesis is given by:
$$h_{\theta}(x) = \theta_0 + \theta_1x + \theta_2x^2 + \theta_3x^3$$

**2. Equivalence with Transformed Feature Map:**
Let's define a feature map $\phi : \mathbb{R} \rightarrow \mathbb{R}^4$ as:
$$\phi(x) = \begin{bmatrix} 1 \\ x \\ x^2 \\ x^3 \end{bmatrix}$$

And let the parameter vector $\theta$ be:
$$\theta = \begin{bmatrix} \theta_0 \\ \theta_1 \\ \theta_2 \\ \theta_3 \end{bmatrix}$$

If we let $\tilde{x} = \phi(x)$, then the linear regression hypothesis $h_{\theta}(\tilde{x}) = \theta^T\tilde{x}$ becomes:
$$h_{\theta}(\tilde{x}) = \begin{bmatrix} \theta_0 & \theta_1 & \theta_2 & \theta_3 \end{bmatrix} \begin{bmatrix} 1 \\ x \\ x^2 \\ x^3 \end{bmatrix} = \theta_0 \cdot 1 + \theta_1 \cdot x + \theta_2 \cdot x^2 + \theta_3 \cdot x^3$$
This is exactly the cubic polynomial hypothesis $h_{\theta}(x)$. Thus, fitting a cubic polynomial is equivalent to fitting a linear regression model on the transformed feature space.

**3. Objective Function $J(\theta)$ for Linear Regression on Transformed Dataset:**
Given the transformed dataset ${(\tilde{x}^{(i)},y^{(i)})}_{i=1}^n$, the Mean Squared Error (MSE) objective function for linear regression is:
$$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta}(\tilde{x}^{(i)}) - y^{(i)})^2$$
Where $m$ is the number of training examples and $h_{\theta}(\tilde{x}^{(i)}) = \theta^T \tilde{x}^{(i)}$.

**4. Batch Gradient Descent Update Rule for $\theta$:**
From Problem 1(a), the batch gradient descent update rule for the parameter vector $\theta$ is:
$$\theta := \theta - \alpha \nabla_{\theta} J(\theta)$$
Where $\nabla_{\theta} J(\theta)$ is the gradient of the cost function with respect to $\theta$. Using the vector form derived previously:
$$\nabla_{\theta} J(\theta) = \frac{1}{m} \tilde{X}^T (\tilde{X}\theta - \vec{y})$$
Here, $\tilde{X}$ is the design matrix where each row is a transformed input $\tilde{x}^{(i)}$, and $\vec{y}$ is the vector of target values. Therefore, the update rule is:
$$\theta := \theta - \alpha \frac{1}{m} \tilde{X}^T (\tilde{X}\theta - \vec{y})$$
For each individual parameter $\theta_j$:
$$\theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_{\theta}(\tilde{x}^{(i)}) - y^{(i)}) \tilde{x}_j^{(i)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

np.seterr(all='raise')

factor = 2.0

class LinearModel(object):
    """Base class for linear models."""

    def __init__(self, theta=None):
        self.theta = theta

    def fit(self, X, y):
        """Solve normal equations: theta = (X^T X)^-1 X^T y"""
        # Using np.linalg.solve for numerical stability as per remark
        self.theta = np.linalg.solve(X.T @ X, X.T @ y)

    def create_poly(self, k, X):
        """Generates polynomial features from x^0 to x^k."""
        # X[:, 1] contains the actual 'x' value (0 column is intercept)
        x = X[:, 1]
        poly_features = [np.ones_like(x)] # x^0
        for i in range(1, k + 1):
            poly_features.append(x**i)
        return np.stack(poly_features, axis=1)

    def create_sin(self, k, X):
        """Generates polynomial features plus a sin(x) term."""
        poly_x = self.create_poly(k, X)
        sin_x = np.sin(X[:, 1]).reshape(-1, 1)
        return np.concatenate([poly_x, sin_x], axis=1)

    def predict(self, X):
        """Make a prediction: h(x) = X * theta"""
        return X @ self.theta

def run_exp(train_path, sine=False, ks=[1, 2, 3, 5, 10, 20], filename='plot.png'):
    train_x, train_y = load_dataset(train_path, add_intercept=True)
    plot_x_raw = np.ones([1000, 2])
    plot_x_raw[:, 1] = np.linspace(-factor*np.pi, factor*np.pi, 1000)

    plt.figure()
    plt.scatter(train_x[:, 1], train_y, color='black', label='Training data', s=10)

    for k in ks:
        model = LinearModel()
        if sine:
            train_features = model.create_sin(k, train_x)
            plot_features = model.create_sin(k, plot_x_raw)
        else:
            train_features = model.create_poly(k, train_x)
            plot_features = model.create_poly(k, plot_x_raw)

        model.fit(train_features, train_y)
        plot_y = model.predict(plot_features)

        plt.ylim(-2, 2)
        plt.plot(plot_x_raw[:, 1], plot_y, label='k=%d' % k)

    plt.legend()
    plt.title(f"{'Sine + Poly' if sine else 'Poly'} Features (ks={ks})")
    plt.savefig(filename)
    plt.show()
    plt.clf()

def main(train_path, small_path, eval_path):
    # Q3: Higher-degree polynomial models
    run_exp(train_path, sine=False, ks=[3], filename='poly_fits.png')

    # Q3: Higher-degree polynomial models
    run_exp(train_path, sine=False, ks=[3, 5, 10, 20], filename='poly_fits.png')

    # Q4: Augmented feature maps (Sine)
    run_exp(train_path, sine=True, ks=[0, 1, 2, 3, 5, 10, 20], filename='sine_fits.png')

    # Q5: Overfitting with limited data
    run_exp(small_path, sine=False, ks=[1, 2, 5, 10, 20, 50], filename='small_poly_fits.png')

if __name__ == '__main__':
    main(train_path='train_data.csv',
         small_path='small_data.csv',
         eval_path='test_data.csv')
```
